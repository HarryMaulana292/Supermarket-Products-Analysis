The columns in this dataset, i.e.

1. **People**

    ● ID: Customer's unique identifier

    ● Year_Birth: Customer's birth year

    ● Education: Customer's education level

    ● Marital_Status: Customer's marital status

    ● MntWines: Amount spent on wine in last 2 years

1. **Promotion**

    ● NumDealsPurchases: Number of purchases made with a discount

    ● AcceptedCmp2: 1 if the customer accepted the offer in the 2nd campaign, 0 otherwise

    ● AcceptedCmp3: 1 if the customer accepted the offer in the 3rd campaign, 0 otherwise

    ● AcceptedCmp4: 1 if the customer accepted the offer in the 4th campaign, 0 otherwise

    ● AcceptedCmp5: 1 if the customer accepted the offer in the 5th campaign, 0 otherwise

    ● Response: 1 if the customer accepted the offer in the last campaign, 0 otherwise

1. **Place**

    ● NumWebPurchases: Number of purchases made through the company’s website

    ● NumCatalogPurchases: Number of purchases made using a catalog

    ● NumStorePurchases: Number of purchases made directly in stores

    ● NumWebVisitsMonth: Number of visits to the company’s website in the last mont

We can see that there are several columns that have outliers. For that, we will try to analyze the outlier data whether it is worth dropping or not.

```python
def iqr(x):
    Q1 = x.quantile(0.25)
    Q3 = x.quantile(0.75)

    IQR = Q3 - Q1
    l_whisker = Q1 - 1.5 * IQR
    u_whisker = Q3 + 1.5 * IQR

    return l_whisker, u_whisker
```

We can also generate the generation column for each customer based on their Year_Birth.

Reporting from genhq.com[[3]](https://genhq.com/the-generations-hub/generational-faqs/), some classifications of generations based on their year of birth are as follows:

1. Pre-Boomer (Born in 1945 and below)
1. Baby Boomer (Born between 1946 - 1964)
1. Gen X (Born between 1965 - 1976)
1. Millennials (Born between 1977 - 1995)
1. Gen Z (Born in 1996 and above)

```python
def gen(x):
    if x <= 1945:
        return "Pre-Boomer"
    if x <= 1964:
        return "Baby Boomer"
    if x <= 1976:
        return "Gen X"
    if x <= 1995:
        return "Millenial"
    else:
        return "Gen Z"
```

```python
df["generation_type"] = df["Year_Birth"].apply(gen)
df.head()
```

Next, the total dependents of each customer will be generated.

```python
df["total_dependents"] = df["Kidhome"] + df["Teenhome"]
df.head()
```

Next, the income group column for each customer will be generated.

Based on the classification of income levels in Texas[[1]](https://demographics.texas.gov/Search?q=texas%20income%20classification), the classification can be divided into 3 levels as follows.

* Upper Level: Income > $75,000 per year
* Middle Level : Income between $25,000 - 75,000 per year
* Lower Level: Income < $25,000 per year


```python
def incomes(x):
    if x > 75000 :
        return "Upper Level"
    elif x >= 25000 :
        return "Middle Level"
    else :
        return "Lower Level"

df["income_class"] = df["Income"].apply(incomes)
df["income_class"].value_counts()
```

Next, we will generate a column of total products that have been purchased by each customer from the supermarket.

```python
df["Total_product_purchased"] = df["wines_products"] + df["fruits_products"] + df["meat_products"] + df["fish_products"] + df["sweet_products"] + df["gold_products"]
df.head()
```
