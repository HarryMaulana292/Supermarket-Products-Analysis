Adapun kolom yang ada pada dataset ini, yaitu :

1. **People**

    ● ID: Customer's unique identifier

    ● Year_Birth: Customer's birth year

    ● Marital_Status: Customer's marital status

    ● Kidhome: Number of children in customer's household

    ● Teenhome: Number of teenagers in customer's household

    ● Dt_Customer: Date of customer's enrollment with the company

    ● Recency: Number of days since customer's last purchase

    ● Complain: 1 if the customer complained in the last 2 years, 0 otherwise

1. **Products**

    ● MntWines: Amount spent on wine in last 2 years

    ● MntFruits: Amount spent on fruits in last 2 years

    ● MntMeatProducts: Amount spent on meat in last 2 years

    ● MntFishProducts: Amount spent on fish in last 2 years

    ● MntSweetProducts: Amount spent on sweets in last 2 years

    ● MntGoldProds: Amount spent on gold in last 2 years

1. **Promotion**

    ● NumDealsPurchases: Number of purchases made with a discount

    ● AcceptedCmp1: 1 if the customer accepted the offer in the 1st campaign, 0 otherwise

    ● AcceptedCmp2: 1 if the customer accepted the offer in the 2nd campaign, 0 otherwise

    ● AcceptedCmp3: 1 if the customer accepted the offer in the 3rd campaign, 0 otherwise

    ● AcceptedCmp4: 1 if the customer accepted the offer in the 4th campaign, 0 otherwise

    ● AcceptedCmp5: 1 if the customer accepted the offer in the 5th campaign, 0 otherwise

    ● Response: 1 if the customer accepted the offer in the last campaign, 0 otherwise

1. **Place**

    ● NumWebPurchases: Number of purchases made through the company’s website

    ● NumCatalogPurchases: Number of purchases made using a catalog

    ● NumStorePurchases: Number of purchases made directly in stores

    ● NumWebVisitsMonth: Number of visits to the company’s website in the last mont

Dapat kita lihat bahwa ada beberapa kolom yang memiliki outlier. Untuk itu kita akan mencoba untuk menganalisa data outlier tersebut apakah layak untuk di drop atau tidak. 

```python
def iqr(x):
    Q1 = x.quantile(0.25)
    Q3 = x.quantile(0.75)

    IQR = Q3 - Q1
    l_whisker = Q1 - 1.5 * IQR
    u_whisker = Q3 + 1.5 * IQR

    return l_whisker, u_whisker
```

Kita juga dapat generate kolom generasi tiap customer berdasarkan Year_Birth nya.

Dilansir dari genhq.com[[3]](https://genhq.com/the-generations-hub/generational-faqs/), beberapa klasifikasi generasi berdasarkan tahun lahirnya adalah sebagai berikut :

1. Pre-Boomer (Kelahiran tahun 1945 ke bawah)
1. Baby Boomer (Kelahiran antara 1946 - 1964)
1. Gen X (Kelahiran antara 1965 - 1976)
1. Millenial (Kelahiran antara 1977 - 1995)
1. Gen Z (Kelahiran tahun 1996 ke atas)

```python
def gen(x):
    if x <= 1945:
        return "Pre-Boomer"
    if x <= 1964:
        return "Baby Boomer"
    if x <= 1976:
        return "Gen X"
    if x <= 1995:
        return "Millenial"
    else:
        return "Gen Z"
```

```python
df["generation_type"] = df["Year_Birth"].apply(gen)
df.head()
```

Berikutnya akan di generate total tanggungan tiap customer.

```python
df["total_dependents"] = df["Kidhome"] + df["Teenhome"]
df.head()
```

Selanjutnya akan di generate kolom kelompok income tiap customer

Berdasarkan klasifikasi tingkat pendapatan di Texas[[1]](https://demographics.texas.gov/Search?q=texas%20income%20classification), klasifikasinya dapat dibagi menjadi 3 tingkatan sebagai berikut.

* Upper Level : Income > $75.000 per tahun
* Middle Level : Income antara $25.000 - 75.000 per tahun
* Lower Level : Income < $25.000 per tahun

```python
def incomes(x):
    if x > 75000 :
        return "Upper Level"
    elif x >= 25000 :
        return "Middle Level"
    else :
        return "Lower Level"

df["income_class"] = df["Income"].apply(incomes)
df["income_class"].value_counts()
```

Selanjutnya akan generate kolom total product yang telah dibeli tiap customer dari supermarket

```python
df["Total_product_purchased"] = df["wines_products"] + df["fruits_products"] + df["meat_products"] + df["fish_products"] + df["sweet_products"] + df["gold_products"]
df.head()
```

### 2. Menambah produk yang memiliki potensi tinggi yang dapat menarik lebih banyak customer. 

Setelah sebelumnya telah kita analisis bagaimana cara meningkatkan penjualan produk yang penjualannya rendah, pada tahap ini kita akan mencoba mencari tahu apa saja produk yang berpotensi untuk menarik customer.

Untuk dapat mengetahui produk apa saja yang berpotensi menarik lebih banyak customer, kita perlu mengetahui proporsi customer supermarket saat ini berdasarkan Marital Status dan Jenis Generasi berdasarkan tahun lahirnya.

```python
df.head()
```

```python
df2 = df["Marital_Status"].value_counts().reset_index().rename(columns = {"Marital_Status" : "Total Customer", "index" : "Marital Status"})
df2
```

```python
plt.figure(figsize = (8,5))
clr = ["maroon", "maroon", "#D8D8D8", "#D8D8D8"]
sns.barplot(data = df2, x = "Marital Status", y = "Total Customer", palette = clr)
plt.title("Total Customer Berdasarkan Marital Status", size = 18, fontweight = "bold",
          y = 1.16, color = "maroon")
plt.grid(alpha = 0.5, color = "maroon")
plt.ylabel("Total Customer", fontweight = "bold", fontsize = 13, fontstyle = "italic")
plt.xlabel("Marital Status", fontweight = "bold", fontsize = 13, fontstyle = "italic")
plt.text(s = "Dua kategori customer berdasarkan marital status terbanyak adalah Married dan Single.\nAnalisis kali ini akan difokuskan kepada kedua marital status terbanyak.",
         x = -0.5, y = 1550)

index1 = (df2["Total Customer"][0])
index2 = (df2["Total Customer"][1])
new_list = [f"{index1:.0f}", f"{index2:.0f}"]

for i in range (len(new_list)):
    plt.annotate(text = new_list[i], xy = (i, df2["Total Customer"][i]/2), size = 15,
                 ha = "center", va = "center", fontweight = "bold", color = "white")
plt.show()
```

* Berdasarkan data, dapat dilihat jika customer supermarket paling banyak adalah mereka yang telah married. Kemudian diurutan kedua terbanyak adalah customer yang masih single.
